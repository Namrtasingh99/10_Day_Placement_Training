```python
import pandas as pd

data = pd.read_excel("Data set.xlsx")
data.head()
```

```python
data.info()
data.describe()
```

```python
data.mean()
data.median()
data.mode()
```

```python
numeric_data = data.select_dtypes(include=['number'])
numeric_data.var()
numeric_data.std()
numeric_data.max() - numeric_data.min()
```

```python
numeric_data = data.select_dtypes(include=['number'])

print("Skewness:\n", numeric_data.skew())
print("Kurtosis:\n", numeric_data.kurtosis())
```

```python
import matplotlib.pyplot as plt

numeric_data.hist(figsize=(15,10))

plt.tight_layout()
plt.show()
```
